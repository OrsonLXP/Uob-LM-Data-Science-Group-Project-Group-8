```
import pandas as pd

# Load the dataset
file_path = "economy table 10.csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info(), df.head()
```

```
# 清理数据：移除千分位逗号并转换为数值类型
for col in df.columns[3:]:  # 从1998年开始的列
    df[col] = df[col].astype(str).str.replace(',', '').astype(float)

# 重新检查数据类型
df.info(), df.describe()
```

```
import matplotlib.pyplot as plt

# 绘制总体时间趋势（按年份均值）
yearly_avg = df.iloc[:, 3:].mean()  # 计算每年的均值

plt.figure(figsize=(12, 6))
plt.plot(yearly_avg.index, yearly_avg.values, marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.title("Overall Trend Over Time (1998-2022)")
plt.grid(True)
plt.show()
```

```
import seaborn as sns

# 绘制直方图
plt.figure(figsize=(12, 6))
sns.histplot(yearly_avg, bins=20, kde=True)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Yearly Average Values")
plt.grid(True)
plt.show()

# 绘制箱线图（Box Plot）
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.iloc[:, 3:])
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Box Plot of Values Over Time")
plt.grid(True)
plt.show()
```

```
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 绘制 ACF 和 PACF 图表
plt.figure(figsize=(12, 6))
plot_acf(yearly_avg, lags=20)
plt.title("Autocorrelation Function (ACF)")
plt.show()

# 重新绘制 PACF，减少滞后阶数（最大 12）
plt.figure(figsize=(12, 6))
plot_pacf(yearly_avg, lags=10)
plt.title("Partial Autocorrelation Function (PACF) - Adjusted")
plt.show()

```

```
from statsmodels.tsa.stattools import adfuller

# ADF 平稳性检验
adf_test = adfuller(yearly_avg)

# 提取 ADF 统计结果
adf_results = {
    "ADF Statistic": adf_test[0],
    "p-value": adf_test[1],
    "Critical Values": adf_test[4],
    "Number of Lags Used": adf_test[2],
    "Number of Observations Used": adf_test[3]
}

adf_results
```

```
# 进行一阶差分
diff_series = yearly_avg.diff().dropna()

# 再次进行 ADF 平稳性检验
adf_test_diff = adfuller(diff_series)
adf_results_diff = {
    "ADF Statistic": adf_test_diff[0],
    "p-value": adf_test_diff[1],
    "Critical Values": adf_test_diff[4],
    "Number of Lags Used": adf_test_diff[2],
    "Number of Observations Used": adf_test_diff[3]
}

# 绘制差分后数据的时序图
plt.figure(figsize=(12, 6))
plt.plot(diff_series, marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Differenced Value")
plt.title("First-order Differenced Series")
plt.grid(True)
plt.show()

adf_results_diff
```

```
# 重新加载数据并执行时间序列建模
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 重新加载数据
file_path = "economy table 10.csv"
df = pd.read_csv(file_path)

# 清理数据（转换年份列为数值型）
for col in df.columns[3:]:  
    df[col] = df[col].astype(str).str.replace(',', '').astype(float)

# 计算每年的均值
yearly_avg = df.iloc[:, 3:].mean()

# 选择 ARIMA(1,1,1) 进行建模
model = ARIMA(yearly_avg.values, order=(1, 1, 1))
model_fit = model.fit()

# 预测未来 5 年
future_steps = 5
forecast = model_fit.forecast(steps=future_steps)

# 生成年份索引
future_years = list(range(2023, 2023 + future_steps))

# 绘制预测结果
plt.figure(figsize=(12, 6))
plt.plot(yearly_avg.index, yearly_avg.values, marker='o', label="Historical Data")
plt.plot(future_years, forecast, marker='o', linestyle="dashed", label="Forecast")
plt.xlabel("Year")
plt.ylabel("Predicted Value")
plt.title("ARIMA(1,1,1) Forecast for Next 5 Years")
plt.legend()
plt.grid(True)
plt.show()

# 预测结果转为 DataFrame 并输出
forecast_df = pd.DataFrame({"Year": future_years, "Forecast": forecast})
print(forecast_df)



```
