### READ TRAFFIC DATA

```python
# traffic data
trafficPath = r'..\datasets\dft_traffic_counts_raw_counts.csv'
traffic = pd.read_csv(datapath.trafficPath,
                      usecols=['year', 'count_date',
                               'region_id', 'region_name', 'region_ons_code', 'local_authority_id', 'local_authority_name', 'local_authority_code',
                               'road_category', 'road_type'
                               'pedal_cycles', 'two_wheeled_motor_vehicles', 'cars_and_taxis', 'buses_and_coaches', 'LGVs',
                               'HGVs_2_rigid_axle', 'HGVs_3_rigid_axle', 'HGVs_4_or_more_rigid_axle', 'HGVs_3_or_4_articulated_axle', 'HGVs_5_articulated_axle', 'HGVs_6_articulated_axle',
                               'all_HGVs','all_motor_vehicles'])
traffic['count_date'] = pd.to_datetime(traffic['count_date'])
print(traffic.shape)
traffic.head()
```

```python
# create month and day columns
traffic.insert(
    traffic.columns.get_loc("count_date") + 1, "month", traffic["count_date"].dt.month
)
traffic.insert(
    traffic.columns.get_loc("count_date") + 2, "day", traffic["count_date"].dt.day
)
```

```python
print(traffic.shape)
traffic.head()
```

```python
# check if there are NaN values in the traffic dataset
nan_count = traffic.isna().sum().sum()
print(f"NaN total: {nan_count}")
nan_per_column = traffic.isna().sum()
nan_per_column = nan_per_column[nan_per_column > 0].sort_values(ascending=False)
print(f"NaN count per column: \n{nan_per_column}")
```

```python
# print each raw with NaN values
nan_rows = traffic[traffic.isna().any(axis=1)] 
nan_rows
```

```python
# fill NaN values with the mean of the columns
traffic_filled = traffic.fillna(traffic.mean(numeric_only=True))
```

```python
# check if there are NaN values in the traffic dataset
nan_count_filled = traffic_filled.isna().sum().sum()
print(f"NaN total after filling: {nan_count_filled}")
```

```python
# save cleaned traffic data
traffic_filled.to_csv(r"..\datasets\traffic_data_cleaned.csv", index=False)
print("CSV file saved successfully!")
```

#### group traffic data & create intensity indicator by month average

```python
numerical_columns = ['pedal_cycles', 'two_wheeled_motor_vehicles', 'cars_and_taxis', 'buses_and_coaches', 
                     'LGVs', 
                     'HGVs_2_rigid_axle', 'HGVs_3_rigid_axle', 'HGVs_4_or_more_rigid_axle', 'HGVs_3_or_4_articulated_axle',
                     'HGVs_5_articulated_axle', 'HGVs_6_articulated_axle',
                     'all_HGVs','all_motor_vehicles']
traffic_monthly = traffic.groupby(['region_name','year', 'month', 'local_authority_code', 'local_authority_name'])\
    [numerical_columns].mean().reset_index().round(3)
print(traffic_monthly.shape)
traffic_monthly.head()
```
