```python
#每年总可支配家庭收入增长率
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取 Excel 文件
file_path = "D:\downloadinternet\Regional gross disposable household income(all ITL).xlsx"
xls = pd.ExcelFile(file_path)

def process_and_plot_table5():
    # 读取 Table 5 数据
    df = pd.read_excel(xls, sheet_name="Table 5")

    # 清理数据
    df_cleaned = df.rename(columns=df.iloc[0]).iloc[1:].reset_index(drop=True)

    # 统一所有列名为字符串，并去除空格
    df_cleaned.columns = [str(c).strip() for c in df_cleaned.columns]

    # **修正年份列名**：将浮点数年份转换为整数字符串
    df_cleaned.columns = [str(int(float(c))) if c.replace('.', '', 1).isdigit() else c for c in df_cleaned.columns]


    # 仅保留 ITL1 级别的数据（ITL 列为 ITL1）
    df_itl1 = df_cleaned[df_cleaned["ITL"] == "ITL1"]

    # 选择 2000-2022 年的数据
    years = [str(year) for year in range(2000, 2002)]
    existing_years = [year for year in years if year in df_itl1.columns]

    # 选择实际存在的年份列
    df_itl1 = df_itl1[["Region name"] + existing_years]


    # 转换数值类型
    df_itl1[existing_years] = df_itl1[existing_years].apply(pd.to_numeric, errors="coerce")

    # 逐年绘制柱状图
    for year in existing_years:
        plt.figure(figsize=(12, 6))
        plt.bar(df_itl1["Region name"], df_itl1[year], color='b', alpha=0.7)
        plt.xlabel("Region Name")
        plt.ylabel("Growth Rate (%)")
        plt.title(f"Annual growth in total GDHI (%) - {year}year")
        plt.xticks(rotation=45, ha="right")
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

# 运行 Table 5 处理和绘图
process_and_plot_table5()
```
