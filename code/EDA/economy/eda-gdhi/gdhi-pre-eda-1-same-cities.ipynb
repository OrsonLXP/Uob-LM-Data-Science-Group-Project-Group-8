```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

# 设定一个相似度阈值，比如 0.8（80%），可以根据实际情况调整
threshold = 0.8

# 保存相似但不完全相同的名称对
similar_names = []

# 遍历 CSV 与 Excel 中独有的名称
for name_csv in only_csv:
    for name_excel in only_excel:
        ratio = similar(name_csv, name_excel)
        if ratio >= threshold:
            similar_names.append((name_csv, name_excel, ratio))

# 输出结果
print("相似但未被识别为相同的名称对（相似度>=0.8）：")
for name_csv, name_excel, ratio in similar_names:
    print(f"CSV: '{name_csv}'  <-->  Excel: '{name_excel}'  相似度: {ratio:.2f}")
```

```python
import pandas as pd
from difflib import SequenceMatcher

def similarity(a, b):
    """计算两个字符串的相似度"""
    return SequenceMatcher(None, a, b).ratio()

# 相似度阈值
threshold = 0.8

# 读取 CSV 文件数据
file_path_csv = r'dft_traffic_counts_raw_counts.csv'
traffic = pd.read_csv(file_path_csv, low_memory=False)
unique_authorities = traffic["local_authority_name"].dropna().unique()

# 读取 Excel 文件中 table1_itl3 表数据
file_path_excel = "Regional gross disposable household income(all ITL).xlsx"
df_GDHI_ITL3 = pd.read_excel(file_path_excel, sheet_name="table1_itl3")
df_GDHI_ITL3.columns = df_GDHI_ITL3.columns.str.strip()
region_names = df_GDHI_ITL3["Region_name"].dropna().unique()

# 将两个名称列表转换为集合
set_csv = set(unique_authorities)
set_excel = set(region_names)

# 精确匹配交集
common_exact = set_csv.intersection(set_excel)

# 剩余未匹配部分
unmatched_csv = set_csv - common_exact
unmatched_excel = set_excel - common_exact

# 使用列表生成式计算相似匹配对（每个只调用一次 similarity）
similar_pairs = [
    (name_csv, name_excel, similarity(name_csv, name_excel))
    for name_csv in unmatched_csv
    for name_excel in unmatched_excel
    if similarity(name_csv, name_excel) >= threshold
]

# 将模糊匹配结果加入共同集合
common_csv = common_exact.union({pair[0] for pair in similar_pairs})
common_excel = common_exact.union({pair[1] for pair in similar_pairs})

# 输出模糊匹配的结果
print("相似但未被精确匹配的名称对（相似度>=0.8）：")
for name_csv, name_excel, ratio in similar_pairs:
    print(f"CSV: '{name_csv}'  <-->  Excel: '{name_excel}'  相似度: {ratio:.2f}")

# 根据共同名称提取对应数据形成新表
df_csv_common = traffic[traffic["local_authority_name"].isin(common_csv)]
df_excel_common = df_GDHI_ITL3[df_GDHI_ITL3["Region_name"].isin(common_excel)]

print("\nCSV 新表行数:", df_csv_common.shape[0])
print("Excel 新表行数:", df_excel_common.shape[0])

# 可选：将新表保存为 CSV 文件
df_csv_common.to_csv("CSV_common_data.csv", index=False)
df_excel_common.to_csv("Excel_common_data.csv", index=False)
```
