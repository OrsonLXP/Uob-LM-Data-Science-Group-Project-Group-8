## road categories: PA, PM, TA, TM
PM: M or Class A Principal Motorway

    解释：主要高速公路（或称 A 类高速公路），通常是连接主要城市、地区的快速公路。

PA: Class A Principal road

    解释：A 类主要公路，主要的干线公路，通常用于交通量较大的地方。

TM: M or Class A Trunk Motorway

    解释：A 类干线高速公路，类似于 PM 类，但具有更高的交通流量或重要性。

TA: Class A Trunk road

    解释：A 类干线公路，连接城市和重要地点，具有较高的运输重要性。

M: Minor road

    解释：小型道路，通常用于低流量、地区性交通的道路。

Of which…: 从 "Minor road"（小型道路）中分离出来的不同类别的道路。

    MB: Class B road
        解释：B 类道路，相较于 A 类公路，流量较小，但仍然是主要的地区连接道路。
    MCU: Class C road or Unclassified road
        解释：C 类道路或未分类道路，通常是最小的道路类型，用于当地的连接或服务功能。
