## Generating eTraffic data with lag, from lag0 to lag4

```python
import os
import pandas as pd

root = os.path.join(os.getcwd(), 'data')
traffic_quarter_path = os.path.join(root, 'final_matched_quarter_traffic_v4.csv')       # traffic_counts_4_filled
```

```python
traffic_quarter_df = pd.read_csv(traffic_quarter_path)
traffic_quarter_df.head()
```

```python
traffic_quarter_df = pd.read_csv(traffic_quarter_path)
def quarter_to_datetime(quarter_str):
    quarter_datetime = pd.to_datetime(quarter_str)
    return pd.Period(quarter_datetime, freq='Q')

traffic_quarter_df['year_quarter'] = traffic_quarter_df['year_quarter'].apply(quarter_to_datetime)
traffic_quarter_df.head()
```

```python
# set lag from 0 ~ 4
lag = 4
traffic_quarter_df['year_quarter_lag'] = traffic_quarter_df['year_quarter'] + lag
traffic_quarter_df[['local_authority_name', 'year_quarter', 'year_quarter_lag']]
```

```python
traffic = traffic_quarter_df.drop(columns=['year_quarter'])
traffic['year'] = traffic['year_quarter_lag'].dt.year
print(traffic.columns)
traffic[['local_authority_name', 'year', 'year_quarter_lag']]
```

```python
columns_to_move = ['local_authority_name', 'year']
remaining_cols = [col for col in traffic.columns if col not in columns_to_move]
traffic = traffic[columns_to_move + remaining_cols]
traffic.head()
```

```python
traffic.drop(columns=['region_id', 'region_name', 'region_ons_code', 'local_authority_id',
       'local_authority_code', 'easting', 'northing', 'latitude', 'longitude',
       'count_records'], inplace=True)
print(traffic.columns)
traffic.head()
```

```python
target_features = traffic.columns.drop(['year_quarter_lag', 'local_authority_name', 'year'])
traffic_grouped = traffic.groupby(['local_authority_name', 'year'])[target_features].sum().reset_index()
traffic_grouped.head()
```

```python
# traffic = pd.read_csv(finalPath)
gdp = pd.read_csv(os.path.join(root, 'GDP_ITL3_LDA(raw).csv'))
print(traffic.shape, gdp.shape)
```

```python
gdp.head()
```

```python
traffic_grouped.head()
```

```python
eTraffic = pd.merge(traffic_grouped, gdp, how='inner', left_on=['local_authority_name', 'year'], right_on=['ITL321NM', 'year'])
print(eTraffic.shape)
eTraffic.head()
```

```python
eTraffic.isnull().any()
```

```python
eTraffic.drop(columns=['LAD23CD', 'LAD23NM', 'ITL321CD', 'ITL321NM'], inplace=True)
eTraffic.drop(columns=['ITL', 'ITL code'], inplace=True)
eTraffic.drop(columns=['Region Name'], inplace=True)
eTraffic.head()
eTraffic.rename(columns={"value": "GDP"}, inplace=True)
eTraffic.shape
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

eTraffic['local_authority_name_encoded'] = encoder.fit_transform(eTraffic['local_authority_name'])
eTraffic.shape
```

```python
if lag == 0:
    eTraffic = eTraffic[eTraffic['year'] < 2020]
    eTraffic = eTraffic[eTraffic['year'] >= 2000]
    print(eTraffic['year'].min(), eTraffic['year'].max())
    print(eTraffic.shape)
elif lag >= 1 or lag <= 4:
    eTraffic = eTraffic[eTraffic['year'] < 2020]
    eTraffic = eTraffic[eTraffic['year'] > 2000]
    print(eTraffic['year'].min(), eTraffic['year'].max())
    print(eTraffic.shape)
```

```python
eTraffic.to_csv(os.path.join(root, f'eTraffic_final_lag{lag}.csv'), index=False)
```
